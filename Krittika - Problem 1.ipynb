```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Dataset_Q1.csv')
print(data)
```

```python
data['Roll Number']
```

```python
roll_list = np.array(data['Roll Number'])
roll_list
```

```python
pandas_dataframe = pd.DataFrame(data)
pandas_dataframe
```

```python
def findLen(str):
    counter = 0    
    for i in str:
        counter += 1
    return counter

lengths = np.zeros(103)
sequence = np.arange(0,103,1)
sequence

for i in sequence:
    lengths[i] = findLen(roll_list[i])
    
print(lengths)
```

```python
pandas_dataframe['Length of Roll Number'] = lengths
pandas_dataframe
```

```python
pandas_dataframe.loc[pandas_dataframe['Length of Roll Number'] != 9]
```

```python
filtered = pandas_dataframe.loc[pandas_dataframe['Length of Roll Number'] != 9]
filtered
```
